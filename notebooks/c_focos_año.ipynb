```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/sucio/rii_d_cant_focos_calor_año.csv', sep=';', quotechar='"')
df.head(12)
```

```python
df.shape
```

```python
df.dtypes
```

```python

df['año'] = pd.to_datetime(df['año'], format='%Y')

```

```python
cantidad_focos_calor= df.groupby('año')['cantidad_focos_de_calor'].sum()

#tamaño de la figura
plt.figure(figsize=(10,5))
#titulo
plt.title("Cantidad de Focos de Calor por Año")
#pasamos los valores x, y
plt.plot(cantidad_focos_calor.index, cantidad_focos_calor.values, marker='o', linestyle='-', color='firebrick')
plt.xlabel("Año")
plt.ylabel("Cantidad de Focos de Calor")
plt.xticks(rotation=20)
plt.grid()
plt.savefig("../web/src/assets/c_focos_año.svg", bbox_inches='tight')
plt.show()
```
