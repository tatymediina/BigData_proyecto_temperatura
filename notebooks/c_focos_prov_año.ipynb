```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/sucio/rii_d_cant_focos_calor_prov_2015_2024_totales.csv', sep=';', quotechar='"')
df.head(25)
```

```python
df = df.replace(' ', 0)
df = df.fillna(0)
```

```python
df.head(25)
```

```python
df = df[df['jurisdicción'] != 'Total']
```

```python
df.columns = ['Jurisdicción', '2015', '2016', '2017', '2018', '2019','2020', '2021', '2022', '2023', '2024']
df.head()
```

```python
df.dtypes
```

```python
df['2016'] = pd.to_numeric(df['2016'], errors='coerce')
```

```python
df.head(25)
```

```python
df.to_csv('../data/limpio/c_focos_prov_año.csv', index=False)
```

```python


df = pd.read_csv("../data/limpio/c_focos_prov_año.csv", index_col="Jurisdicción").transpose()
df.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title("Focos de Calor por Jurisdicción y Año")
plt.xlabel("Año")
plt.ylabel("Cantidad de Focos de Calor")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("../web/src/assets/c_focos_prov_año.svg", bbox_inches='tight')
plt.show()
```
