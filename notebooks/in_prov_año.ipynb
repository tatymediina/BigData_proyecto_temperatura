```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/sucio/rii_b_cantidad_incendios_prov_2017_2025.csv", sep=';', quotechar='"')
df.head()
```

```python
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df.columns = ['Jurisdicción', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025']
```

```python
df.head(25)
```

```python
df_copy = df.copy()
```

```python
df = df[df['Jurisdicción'] != 'Total']
```

```python
import numpy as np

df.replace("s/d", np.nan, inplace=True)
```

```python
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(25)
```

```python
df = df[['Jurisdicción', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']]
```

```python
df.to_csv("../data/limpio/in_prov_año.csv", index=False)

```

```python

df = pd.read_csv("../data/limpio/in_prov_año.csv", index_col="Jurisdicción").transpose()
plt.figure(figsize=(14, 8))
for provincia in df.columns:
    plt.plot(df.index, df[provincia], marker='o', label=provincia, linewidth=2)

plt.title("Evolución de Incendios Reportados por Jurisdicción", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Cantidad de Incendios", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../web/src/assets/in_prov_año.svg", bbox_inches='tight')
plt.show()
```

```python
top_n = 10
top_jurisdicciones = df.iloc[-2].sort_values(ascending=False).head(top_n).index


df_filtrado = df[top_jurisdicciones]


plt.figure(figsize=(12, 6))
for provincia in df_filtrado.columns:
    plt.plot(df_filtrado.index, df_filtrado[provincia], marker='o', label=provincia, linewidth=2)

plt.title(f"Top {top_n} Evolución de Incendios por Jurisdiccion con Mayor Cantidad de Incendios (2017-2025)", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Cantidad", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
