```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/sucio/rii_c_hectareas_incendios_prov_2017_2025.csv", sep=';', quotechar='"')
df.head(25)
```

```python
print(df.shape)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python

df = df[df['jurisdicción'] != 'Total']
```

```python
import numpy as np

df.replace("s/d", np.nan, inplace=True)
```

```python
# Reemplazamos todos los NaN con 0 (si sabemos que significa "no hubo incendios")
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(25)
```

```python
df.columns = ['Jurisdicción', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025']
```

```python
df.head(25)
```

```python
# Convertimos todas las columnas numéricas (menos la de jurisdicción)
columnas_numericas = df.columns[1:] 

for col in columnas_numericas:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.dtypes
```

```python
import seaborn as sns

# Pasamos el dataframe a formato long para facilitar el heatmap
df_melted = df.melt(id_vars='Jurisdicción', var_name='Año', value_name='Hectáreas')


pivot = df_melted.pivot(index='Jurisdicción', columns='Año', values='Hectáreas')

plt.figure(figsize=(14, 10))
sns.heatmap(pivot, cmap='YlOrRd', annot=False, linewidths=0.5)

plt.title('Heatmap de Superficie Afectada por Incendios')
plt.xlabel('Año')
plt.ylabel('Jurisdicción')
plt.tight_layout()
plt.savefig("../web/src/assets/c_hect_prov_año.svg", bbox_inches='tight')
plt.show()
```
