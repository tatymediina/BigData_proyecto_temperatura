```python
import matplotlib.pyplot as plt


# Seleccionar solo las columnas necesarias
df_filtrado = df[['Mes', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']]

# Configurar la columna "Mes" como índice
df_filtrado.set_index('Mes', inplace=True)

# Graficar
df_filtrado.plot(figsize=(15,6), marker='o')
plt.title("Evolución Mensual de Incendios Reportados")
plt.xlabel("Mes")
plt.ylabel("Cantidad de Incendios")
plt.xticks(rotation=45)
plt.grid()
plt.legend(title="Año")
plt.tight_layout()
plt.savefig("../web/src/assets/c_incendios_mes_año.svg", bbox_inches='tight')
plt.show()


```
